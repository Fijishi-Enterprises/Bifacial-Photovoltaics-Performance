## Create an array for days in a year and then download each of the day files from the NOAA website
There are 365 text files that are downloaded, one for each day. These are downloaded as text files and later need to be organized

```python
days = []
values = list(range(1,366))
for v in values:
    days.append(str(v).zfill(3))
```

```python
for day in days:
    URL = "https://gml.noaa.gov/aftp/data/radiation/baseline/spo/2021/spo21"+day+".dat"
    response = requests.get(URL)
    name = r'C:\Users\jjones\Documents\GitHub\bifacial_radiance\bifacial_radiance\TEMP\South_Pole_Weather_Data\spo21'+day+'.txt'
    open(name, "wb").write(response.content)

    
```

## Create Dataframe combining South Pole information
The rows in the tect files are seperated by minute. The below code creates a new dataframe to hold specific information from 
each text file.  They are then averaged for the hour and the rows for hourly data are appended to the dataframe.
Ultimately this creates a data frame containing information for the whole year seperated by hour of the day.


```python
SPole = pd.DataFrame()
for day in days:
    dates = pd.DataFrame()
    data = pd.DataFrame()
    file = r'C:\Users\jjones\Documents\GitHub\bifacial_radiance\bifacial_radiance\TEMP\South_Pole_Weather_Data\spo21'+day+'.txt'
    #df = pd.read_csv(file, sep='\s+', header=None)
    df = pd.read_table(file, delimiter='\s+',skiprows=2, header=None)
    dates['year'] = df[0]
    dates['month'] = df[2]
    dates['day'] = df[3]
    dates['hour'] = df[4]
    #dates['min'] = df[5]
    date = pd.to_datetime(dates)
    date2 = pd.DatetimeIndex(dates)
    #data['date'] =date
    data['DNI'] = df[12]
    data['DHI'] = df[14]
    data['GHI'] = df[8]
    data['windir'] = df[44]
    data['windspeed'] = df[42]
    data['rel_hum'] = df[40]
    data['albedo'] = df[10]/df[8]
    data.index=(pd.to_datetime(date))
    data2 = data.resample('H').mean()
    SPole = SPole.append(data2)

    #SPole = SPole.append({'DNI': data2['DNI'], 'DHI': data2['DHI'], 'GHI': data2['GHI'],'windir': data2['windir'],'windspeed': data2['windspeed'],'rel_hum': data2['rel_hum'],'albedo': data2['albedo']}, ignore_index=True)
    #data3 = data.groupby([pd.Grouper(freq='H')]).sum().T
```
